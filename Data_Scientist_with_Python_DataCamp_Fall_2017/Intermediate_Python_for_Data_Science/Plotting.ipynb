# Matplotlib

From all the visualization packages in Python, **Matplotlib** is the best. The subpage `pyplot` is imported as `plt` by convention. This package will allow you to visualize data in Python in many different types of plots. 

## Line chart
When using the function `plt.plot`, the first argument is the x axis and the second the y axis. This kind of plot is useful when there is a time scale along the horizontal axis.
Python won't show the plot until the `plt.show()` function is executed. This is done so that you can add more elements (e.g., titles, labels, etc.) to the plot before visualizing it.

```python
import matplotlib.pyplot as plt

years = [1950, 1970, 1990, 2010]
population = [2.567, 3.657, 5.678, 6.978]

plt.plot(years, population)
plt.show()
```

# Scatter plots

The `plt.scatter` function behaves very similarly to the line chart one, except it doesn't connect the data points with a line. This is a more "honest" way of visualizing the data because you can tell how many data points there are. You can change the scale of the axis with `plt.xscale("log")`.

```python
plt.scatter(years, population)
plt.show()
```

## Histograms

Very useful way to explore your data, especially their distribution. Basically it calculates how many data points are within each of the bins that you specified; the taller the bar the most data points in that bin. The first argument is a **list** and the second one the number of bins (10 by default). The number of bins is very important, too few and you oversimplify the data, too many and you over complicate it and it's hard to see the bigger picture. The function `plt.clf` clears the current plot, otherwise you plot on top.

```python
#help(plt.hist)

# generate data
import numpy as np
values = np.random.normal(7, 3, 12)

plt.hist(values, bins = 3)
plt.show()
```

## Customization 

Each plot is highly customizable with different colors, shapes, labels, axis, etc. There are many functions including: `plt.xlabel`, `plt.title`, `plt.ytick` (starts the axis from a value, it can also have a second argument to change the names that are displayed on the axis, this **must** be a list of the same length as the first argument.)

It's a good idea to write the labels, title, and other strings into variable, that way you only need to type it once!
Do this, `xlab = "Years"` and then `plt.xlabel(xlab)`.


```python
years = [1950, 1970, 1990, 2010]
population = [2.567, 3.657, 5.678, 6.978]

years = [1800, 1850, 1900] + years
population = [1.0, 1.236, 1.650] + population

plt.plot(years, population)

# Add axis labels
# Using a variable of type string
xlab = "Years"
plt.xlabel(xlab)
# Writting the string 
plt.ylabel("Population")
# Add a title for the plot
plt.title("World Population Projections")
# Starts Y axis at zero and adds names to the values on the axis. Must be lists!
plt.yticks([0, 2, 4, 6, 8, 10], ["2B", "4B", "6B", "8B", "10B"])
plt.show()
```

There are many ways to customize the plot, some of them are `c` for colors, `s` for the scalar (magnitude), `alpha` for opacity, etc. You can also add a grid with `plt.grid()` and text in the plot with `plt.text()` (needs coordinates).

```python
plt.scatter(years, population, c = "blue", alpha = 0.5, s = 500)
plt.grid()
plt.text(1900, 5, "Additinal text!")

plt.show()
```
