# Random Numbers

Sometimes is easier to simulate a scenario thousands of times than it is to analytically calculate the probability of something. For simulations, you need to create your own data and that's where **Random Generators** come in handy. Computers usually generate **_pseudorandom_** numbers that start from a _seed_ that you can specify or not. The numbers that are generated are random but _are_ consistent between runs, i.e., if the seed is the same you will get the same random numbers when you generate them in two separate occasions. This is very useful because it assures reproducibility! Two useful function of Numpy are `rand()` and `randint()`, which generate float and integers respectively. You can specify the range of numbers that you want generated with `randint(1, 5)`, this will generate number from 1 to 4, **5 is not included!**

````python
import numpy as np

np.random.rand()

np.random.seed(123)
print(np.random.rand())
print(np.random.rand())
print("\n")


np.random.seed(123)
print(np.random.rand())
print(np.random.rand())
print("\n")

# Coin toss
# only 0 or 1
coin = np.random.randint(0, 2)
if coin == 0:
    print("Heads")
else:
    print("Tails")

````

Here is a little example of using random numbers (**random walk**) to move up or down depending on their values. `If dice` is 1 or 2, you go one step down. `if dice` is 3, 4 or 5, you go one step up. `Else`, you throw the dice again. The number of eyes is the number of steps you go up.

````python
# Import numpy and set seed
import numpy as np
np.random.seed(123)

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1, 7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <=5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
````

If you want to have a variable be _always_ bigger than a value, use the function `max` and pass it two arguments. The largest will be return. For example, the variable `x` will never be smaller than 10 here, because if it is, 10 will be returned:  
```if something:
    x = (10, x-1)```

To generate a random walk, you need to keep track of how many steps you have taken and in what direction. You can also plot is easily:

````python
# Import numpy and set seed
import numpy as np
np.random.seed(123)

tails = [0]
for x in range(100):
    coin = np.random.randint(0, 2)
    tails.append(tails[x] + coin)
print(tails)

import matplotlib.pyplot as plt
plt.plot(tails)
plt.show()
````

If you generate many random walks, you are going to end up with many results. This will give you a distribution of results that you can use to calculate the probability of ending somewhere.

````python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)

final_tails = [0]
for x in range(10000):
    tails = [0]
    for x in range(10):
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1]) # append the last value of tails, i.e., the total number of tails

#print(final_tails) # prints the total number of tails in each iteration of the inner for loop

plt.hist(final_tails, bins = 10)
plt.show()
````

This is a similar example but it transforms the list of lists (`final_tails`) into a numpy array for better plotting. We also added an `if` clause to simulate "clumsiness", so if a random number is less or equal to 0.0001, you'll fall and have to start again

````python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(123)
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        # Implement clumsiness
        if np.random.rand() <= 0.001:
            step =0
            
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)
# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
# Now every row in np_all_walks represents the position after 1 throw for the 10 random walks.
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()

````
